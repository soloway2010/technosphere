## 1 этап. Сбор данных и их обработка
В ходе поиска оказалось, что невозможно получить доступ к данным о статусе отношенийс точностью до человека. В следствие этого мы решили взять данные всемирной переписи населения по распределению marital status. В данных приведена информация о статусе в 221 стране и разделена по возросту, полу и году опроса. Было решено взять данные только за 2010 год

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from itertools import izip
```

```python
data_Raw = pd.read_csv("./data_Raw.csv", sep=',')
data_Raw.head()
```

```python
data = data_Raw.iloc[:, 0:18]

data = data.drop(["ISO code", "Non-standard age groups"], axis=1);

data = data[data.iloc[:, 1] == "Latest"]

data = data.drop(["Period", "Year"], axis=1)

data["Marital status"] = data["Marital status"].replace("Divorced/Separated", "Divorced")

data = data[(data.iloc[:, 2] == "Single") | (data.iloc[:, 2] == "Married") | (data.iloc[:, 2] == "Divorced")]

data = data[data.iloc[:, 1] != "Both Sexes"]

data = data.dropna()

data = data.reset_index(drop=True)

data.to_csv("./data.csv", sep=',')
```

Так как данные необходимы о каждом человеке в отдельности, потребоволось сгенерировать их самим на основе упомянутых выше данных

```python
data = pd.read_csv("./data.csv", sep=',')

df = pd.DataFrame(columns=["Country", "Sex", "Age", "Marital status"])

unique_countries = np.unique(data.iloc[:, 1])

N = 100000

df["Country"] = np.random.choice(unique_countries, N)
df["Sex"] = np.random.choice(["Men", "Women"], N)
df["Age"] = np.random.choice(range(15, 100), N)

statuses = {}

for i in xrange(N):
    tmp = data[(data["Country"] == df["Country"][i]) & (data["Sex"] == df["Sex"][i])]
    age_group = df["Age"][i] / 5 + 1;
    if age_group > 14:
        age_group = 14

    probs = list(tmp.iloc[:, age_group] / 100)
    probs[0] += 1 - sum(probs)
    statuses[i] = np.random.choice(tmp["Marital status"], 1, p=probs)[0]

df["Marital status"] = list(statuses.values())

df.to_csv("./people.csv", sep=',')
```

```python
data = pd.read_csv("./people.csv", sep=',')
data = data.iloc[:, 1:]
print data.shape
data.head()
```

Как видно, было сгенерировано 100 тыс. экземпяров

```python
data["Marital status"] = data["Marital status"].replace(["Single", "Married", "Divorced"], [0, 1, 2])
data["Sex"] = data["Sex"].replace(["Men", "Women"], [0, 1])
```

```python
countries = np.unique(data["Country"])
i = iter(countries)
to_int_encoder = dict(izip(i, xrange(len(countries))))

country_column = list(data["Country"])
country_column = map(lambda x: [to_int_encoder[x]], country_column)

enc = OneHotEncoder()
new_country_column = enc.fit_transform(country_column).toarray()

data = data.drop(["Country"], axis=1)
```

```python
for i in xrange(new_country_column.shape[1]):
    data["Country feat. #%d" % i] = new_country_column[:, i]
```

## 2 этап. Выбор модели и подбор гиперпараметров

```python
X = data.drop(["Marital status"], axis=1).as_matrix()
y = data["Marital status"].as_matrix()
```

```python
X -= np.mean(X, axis=0)
X /= np.std(X, axis=0)
X
```

```python
clf = DecisionTreeClassifier()
cross_val_score(clf, X[:10000, :], y[:10000], cv=10)
```

```python
clf = LogisticRegression()
cross_val_score(clf, X, y, cv=10)
```

Значительной разницы нет, но логистическая регрессия работает быстрее
